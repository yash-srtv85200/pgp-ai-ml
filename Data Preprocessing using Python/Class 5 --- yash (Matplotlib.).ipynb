```python
x = np.random.rand(100)
y = 2*x + np.random.rand(100)

coefficients = np.polyfit(x, y, 1)
trend_line = np.poly1d(coefficients)

plt.scatter(x, y, label='Random_Data')

# Plot trend line
plt.plot(x, trend_line(x), color='red', label='Trend Line')

# Find max and min points
max_x = x[np.argmax(y)]
max_y = y.max()
min_x = x[np.argmin(y)]
min_y = y.min()

# Label max and min points
plt.text(max_x, max_y, f'Max ({max_x:.2f}, {max_y:.2f})', fontsize=10, ha='right')
plt.text(min_x, min_y, f'Min ({min_x:.2f}, {min_y:.2f})', fontsize=10, ha='right')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Set Seaborn style
sns.set(style='whitegrid')

# Create scatter plot with customizations
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips, color='skyblue', alpha=0.7, edgecolor='k', s=100)

# Set labels and title
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Tip ($)', fontsize=14)
plt.title('Scatter Plot of Total Bill vs Tip', fontsize=16)

# Add a regression line
sns.regplot(x='total_bill', y='tip', data=tips, scatter=False, color='red')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Set Seaborn style
sns.set(style='whitegrid')

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create scatter plot with customizations
plt.figure(figsize=(8, 6))

# Define colors based on the 'sex' column
colors = {'Male': 'blue', 'Female': 'orange'}

# Iterate over unique values in 'sex' column and plot scatter points with different colors
for sex in tips['sex'].unique():
    sns.scatterplot(x='total_bill', y='tip', data=tips[tips['sex'] == sex], 
                    color=colors[sex], alpha=0.7, edgecolor='k', s=100, label=sex)

# Set labels and title
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Tip ($)', fontsize=14)
plt.title('Scatter Plot of Total Bill vs Tip', fontsize=16)

# Add a legend
plt.legend(title='Sex', title_fontsize='14', fontsize='12')

# Add a regression line
sns.regplot(x='total_bill', y='tip', data=tips, scatter=False, color='red')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Create an interactive scatter plot using Plotly Express
fig = px.scatter(tips, x='total_bill', y='tip', color='sex',
                 color_discrete_map={'Male': 'blue', 'Female': 'orange'},
                 title='Total Bill vs Tip',
                 labels={'total_bill': 'Total Bill ($)', 'tip': 'Tip ($)'},
                 hover_name='day', hover_data={'sex': True, 'smoker': True},
                 trendline='ols', trendline_color_override='red')

# Update axis ranges for better visibility
fig.update_xaxes(range=[0, tips['total_bill'].max()*1.1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")
# Set the style using Seaborn
sns.set(style="whitegrid")
# Create a scatter plot with custom style
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x="total_bill", y="tip", hue="sex", size="size", data=tips, palette="viridis", sizes=(20, 200))
# Set plot title and labels
plt.title('Scatter Plot of Total Bill vs Tip', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Tip ($)', fontsize=14)
# Add a legend with a title
legend = scatter.legend(title='Gender', title_fontsize='14', loc='upper right')
legend.get_title().set_fontsize('16')
# Customize the grid and background color
sns.despine()
plt.grid(True, linestyle='--', alpha=0.7)
# Show the plot
plt.show()
```

```python
# Radar Chart
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for a radar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
data = np.random.randint(1, 10, len(categories))

# Create a radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)
data = np.concatenate((data, [data[0]]))
angles = np.concatenate((angles, [angles[0]]))
plt.polar(angles, data, marker='o')

# Add labels
plt.thetagrids(np.degrees(angles[:-1]), labels=categories)

# Show the plot
plt.title('Radar chart: Random data')
plt.show()
```

```python
# 3D
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Add a color bar
cbar = fig.colorbar(ax.plot_surface(x, y, z, cmap='viridis'), ax=ax, pad=0.1)

# Show the plot
plt.show()
```
